```python
# Finding Optimal Threshold
```

```python
from sklearn.metrics import roc_curve

# Step 1: Get class 1 probabilities
y_prob = reg.predict_proba(x_test)[:, 1]

# Step 2: Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Step 3: Compute Youden's J
youden_j = tpr - fpr

# Step 4: Find the best threshold
optimal_idx = np.argmax(youden_j)
optimal_threshold = thresholds[optimal_idx]

print(f"🔍 Optimal Threshold (Youden’s J): {optimal_threshold:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.plot(fpr, tpr, label="ROC Curve")
plt.scatter(fpr[optimal_idx], tpr[optimal_idx], color='red', label=f"Best Threshold: {optimal_threshold:.2f}")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve with Optimal Threshold")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import f1_score

f1_scores = [f1_score(y_test, y_prob >= t) for t in thresholds]
best_f1_idx = np.argmax(f1_scores)
best_f1_threshold = thresholds[best_f1_idx]

print(f"🔍 Optimal Threshold (Max F1): {best_f1_threshold:.4f}")
```
